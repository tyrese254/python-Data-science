```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('/content/drive/MyDrive/account.csv')

print(data.head())

print(data.info())

plt.scatter(data['Weather_Conditions'], data['Number_of_Casualties'])
plt.title('Weather_Conditions vs Number_of_Casualties')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Casualties')
plt.show()

model = LinearRegression()

X = data[['Weather_Conditions']]
y = data['Number_of_Casualties']

model.fit(X, y)

plt.scatter(data['Weather_Conditions'], data['Number_of_Casualties'], label='Data')
plt.title('Weather_Conditions vs Number_of_Casualties')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Casualties')

plt.plot(X, model.predict(X), color='red', label='Linear Regression')
plt.legend()
plt.show()

intercept = model.intercept_
slope = model.coef_[0]

print(f"Regression Equation: Number_of_Casualties = {slope:.2f} * Weather_Conditions + {intercept:.2f}")

plt.scatter(data['Weather_Conditions'], data['Number_of_Casualties'], label='Data')
plt.title('Weather_Conditions vs Number_of_Casualties')
plt.xlabel('Weather_Conditions')
plt.ylabel('Number_of_Casualties')

plt.plot(X, model.predict(X), color='red', label=f'Linear Regression: Number_of_Casualties = {slope:.2f} * Weather_Conditions + {intercept:.2f}')
plt.legend()
plt.show()

persons_involved = 75
predicted_fatalities = model.predict([[persons_involved]])
print(f"Predicted Fatalities for {persons_involved} persons involved in accidents: {predicted_fatalities[0]}")
```
